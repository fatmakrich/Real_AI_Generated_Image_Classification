```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Charger votre mod√®le YOLO personnalis√©
model = YOLO('yolo11n.pt')  # Remplacez par le chemin du mod√®le YOLOv11 si c'est un fichier personnalis√©


# Entra√Æner le mod√®le sur votre dataset
results = model.train(
    data='/kaggle/input/car-damage/data.yaml',  # Chemin vers le fichier de configuration YAML
    epochs=100,        # Nombre d'√©poques
    imgsz=640,         # Taille des images
    batch=16,          # Taille de batch
    plots=True         # G√©n√©rer automatiquement des visualisations pendant l'entra√Ænement
)

# Les r√©sultats seront sauvegard√©s dans le dossier "runs/train/exp/"
```

```python
import os
import shutil

# Chemin du meilleur mod√®le ou du dernier mod√®le
best_model_path = "/kaggle/working/runs/detect/train3/weights/best.pt"
last_model_path = "/kaggle/working/runs/detect/train3/weights/last.pt"

# Cr√©er un r√©pertoire pour enregistrer les mod√®les sauvegard√©s dans /kaggle/working/
save_dir = "/kaggle/working/saved_model/"
os.makedirs(save_dir, exist_ok=True)

# Copier les poids du mod√®le sauvegard√© vers /kaggle/working/
shutil.copy(best_model_path, os.path.join(save_dir, "best_model.pt"))
shutil.copy(last_model_path, os.path.join(save_dir, "last_model.pt"))

# Afficher un message pour confirmer la sauvegarde
print(f"Les mod√®les ont √©t√© sauvegard√©s dans {save_dir}")
```

```python
import os

# Directory containing test images
test_images_directory = '/kaggle/input/test-damage'  # Replace with your test images directory

# Loop through all images in the directory
for image_file in os.listdir(test_images_directory):
    image_path = os.path.join(test_images_directory, image_file)
    
    # Run inference
    results = model(image_path)
    
    # Results contain a list, access the first item for the prediction details
    prediction = results[0]
    
    # Show the results
    prediction.show()  # This will display the image with predictions
```
